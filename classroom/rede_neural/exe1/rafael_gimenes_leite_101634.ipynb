```python
#CÉLULA MLP-MAIN-01

#Parâmetros
#1 - Configuração dos parametros de treinamento
#IMPLEMENTE O CÓDIGO AQUI
eta        = 0.1
epsilon    = 0.000001
nInputs    = 3
nHidden    = 10
nOutputs   = 1
max_epochs = 10000
dictLayers = {'nLayers':2, 'vLayers':np.array([nInputs, nHidden, nOutputs])}

#2 - Carga dos dados de treinamento
#IMPLEMENTE O CÓDIGO AQUI
dfTrainingFile =  model.loadDataFile('ds_resonance_training.csv')

nSamples = len (dfTrainingFile['x1'])

#3 - Separação dos dados de treinamento em entradas e saida
#IMPLEMENTE O CÓDIGO AQUI
vInputsTrain  = np.array(dfTrainingFile[['x1', 'x2', 'x3']])
vOutputsTrain = np.array(dfTrainingFile[['d']])

#4 - Execução do treinamento
#IMPLEMENTE O CÓDIGO AQUI
resultOnline = []
for i in range(5):
    (W_1, W_2, B_1, B_2, MSE) = model.mlp_train_online(eta, epsilon, vInputsTrain.T, vOutputsTrain.T, dictLayers, max_epochs)
    resultOnline.append((W_1, W_2, B_1, B_2, MSE, len(MSE)))
    print('----------------------------------------------------')

from operator import itemgetter 
resultOnline_ordem = sorted(resultOnline, key = itemgetter(5), reverse=True) 

print(W_1)
print('-----')
print(W_2)

#5 - Gráfico do Erro Quadratico Medio
#plt.figure()
#plt.plot(range(len(MSE)-1), MSE[1:])
```

### 7. Explique se é possível realizar o treinamento da rede PMC, por meio do algoritmo backpropagation, quando se inicializa todas as matrizes de pesos com elementos nulos. Discorra também se há então alguma implicação quando se inicializa todos os elementos das matrizes de pesos com valores iguais (diferentes de zeros).

**RESPOSTA:**

Backpropagation envolve a adição aos pesos existentes, não a multiplicação. A quantidade que você adiciona é especificada pela regra delta. Então backpropagation identifica o conjunto de pesos que minimiza a diferença quadrática ponderada entre o alvo e os valores observados

Então existem pelo menos duas boas razões para não definir os pesos iniciais como nulos(zero):

Primeiro, as redes neurais tendem a ficar presas em locais mínimos, então é uma boa ideia dar a elas muitos valores iniciais diferentes. Você não pode fazer isso se todos começarem no zero.

Segundo, se os neurônios começam com os mesmos pesos, então todos os neurônios seguirão o mesmo gradiente e sempre acabarão fazendo a mesma coisa um com o outro.
