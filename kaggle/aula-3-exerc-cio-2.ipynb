# Gráfico de Linha
#### Dataset "anv", váriaveis: Ano Fabricação 


```python
#grafico de linha
#embraer
grp3 = anv[ ((anv['aeronave_ano_fabricacao']>=1990) & (anv['aeronave_fabricante']=='EMBRAER'))]
grp3 = grp3.groupby(['aeronave_ano_fabricacao']).size().nlargest(100).to_frame().reset_index()
grp3.columns = ['Ano','Qtde']
grp3 = grp3.sort_values('Ano')

#BOEING COMPANY
grp4 = anv[ ((anv['aeronave_ano_fabricacao']>=1990) & (anv['aeronave_fabricante']=='BOEING COMPANY'))]
grp4 = grp4.groupby(['aeronave_ano_fabricacao']).size().nlargest(100).to_frame().reset_index()
grp4.columns = ['Ano','Qtde']
grp4 = grp4.sort_values('Ano')



# embraer
trace1 = go.Scatter(
                    x = grp3.Ano,
                    y = grp3.Qtde,
                    mode = "lines",
                    name = "EMBRAER",
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                    text= grp3.Ano)
#boeing
trace2 = go.Scatter(
                    x = grp4.Ano,
                    y = grp4.Qtde,
                    mode = "lines",
                    name = "BOEING COMPANY",
                    marker = dict(color = 'rgba(16, 25, 88, 0.8)'),
                    text= grp4.Ano)
data = [trace1,trace2]
lay = {
  'title': 'Comparação quantidade de Fabricação desde 1990\n entre EMBRAER e BOEING'
};
fig = dict(data = data,layout=lay)
iplot(fig)
```

# Gráfico de Barras
#### Dataset Anv, váriaveis: aeronave_nivel_dano(Dano x Qtde)



```python
#gráfico de barras
tab_aeronave_nivel_dano = anv["aeronave_nivel_dano"].value_counts()
tab_aeronave_nivel_dano = tab_aeronave_nivel_dano.to_frame().reset_index()
tab_aeronave_nivel_dano.columns = ['Dano','Qtde']
tab_aeronave_nivel_dano

# create trace1 
trace1 = go.Bar(
                x = tab_aeronave_nivel_dano.Dano,
                y = tab_aeronave_nivel_dano.Qtde,
                name = "Dano",
                marker = dict(color = 'rgba(0, 20, 155, 0.5)',
                             line=dict(color='rgb(0,0,0)',width=1.5)),
                text = tab_aeronave_nivel_dano.Dano)

data = [trace1]
fig = go.Figure(data = data)
iplot(fig)
```

# Histograma
#### Dataset "cars", váriaveis: horse power, torque




```python
#histograma
#cars2 = cars[((cars['horsepower']<350) & (cars['torque']<350))]
trace0 = go.Histogram(x=cars['horsepower'],name='Horse Power', marker=dict(color='rgba(241, 17, 189, 0.6)'),opacity=0.75)
trace1 = go.Histogram(x=cars['torque'],name='Torque', marker=dict(color='rgba(235, 255, 55, 0.6)'),opacity=0.75)
lay = go.Layout(barmode='overlay',
                   title='Horse Power and Torque',)
data = [trace0,trace1]
fig = go.Figure(data = data,layout=lay)
iplot(fig)
```

# Gráfico de Radar
#### Dataset "pokemon", váriaveis: todas de força



```python
#grafico Radar
x = poke[poke["Name"] == "Oddish"]
oddish = go.Scatterpolar(
  r = [x['HP'].values[0],x['Attack'].values[0],x['Defense'].values[0],x['Sp. Atk'].values[0],x['Sp. Def'].values[0],x['Speed'].values[0],x["HP"].values[0]],
  theta = ['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed','HP'],
  fill = 'toself',name='Oddish',line =  dict(color = 'blue')
)
x = poke[poke["Name"] == "Pikachu"]
pikachu = go.Scatterpolar(
  r = [x['HP'].values[0],x['Attack'].values[0],x['Defense'].values[0],x['Sp. Atk'].values[0],x['Sp. Def'].values[0],x['Speed'].values[0],x["HP"].values[0]],
  theta = ['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed','HP'],
  fill = 'toself', name='Pikachu',line =  dict(color = 'red')
)
data = [oddish,pikachu]

layout = go.Layout(
  polar = dict(
    radialaxis = dict(
      visible = True,
      range = [0, 200]
    )
  ),
  showlegend = False,
  title = "Oddish X Pikachu"
)
fig = go.Figure(data=data, layout=layout)

iplot(fig, filename = "Single Pokemon stats")
```

# Gráfico de Caixa
#### Dataset "sales", váriaveis: Team, Sale



```python
#grafico de caixa
sales.head()

s_alpha = sales[sales.Team== 'Alfa']
s_beta = sales[sales.Team== 'Beta']

trace0 = go.Box(
    y=s_alpha.Sale,
    name = 'Alpha Sale',
    marker = dict(
        color = 'rgb(12, 12, 140)',
    )
)
trace1 = go.Box(
    y=s_beta.Sale,
    name = 'Beta Sale',
    marker = dict(
        color = 'rgb(12, 128, 128)',
    )
)
data = [trace0, trace1]
iplot(data)
```
