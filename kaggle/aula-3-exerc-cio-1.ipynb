# Questão 1

Construa um gráfico de violino que compara o número de eleitores do gênero masculino e gênero feminino das 15 cidades com o maior número de eleitores do estado da Bahia.

```python
df1 = df[(df['uf'] == 'BA')]
df1=df1.nlargest(15,'total_eleitores')

plt.violinplot([df1['total_eleitores'],df1['gen_masculino'],df1['gen_feminino']])
plt.xticks([0,1,2,3], ('','Tot', 'Masc','Fem',''))
plt.show()
```

# Questão 2

```python
df2 = df
df2=df.nlargest(15,'total_eleitores')

plt.violinplot([df2['total_eleitores'],df2['f_18_20'],df2['f_21_24'],df2['f_70_79'],df2['f_sup_79']])
plt.xticks([0,1,2,3,4,5], ('','Tot', 'f_18_20','f_21_24','f_70_79','f_sup_79'))
plt.show()
```
